```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Visualizaciones 

```python
path = 'antropometria-dataset.csv'
df = pd.read_csv(path)
df.head()
```

```python
sns.distplot(df['peso'])
plt.show()
```

```python
df['peso'].mean()
```

```python
df['peso'].describe()
```

```python
sns.displot(df['peso']).set(title='Histograma de peso')
plt.show()
```

```python
sns.displot(df['peso'], kde=True, color='#71DB6B')
plt.show()
```

```python
plt.hist(df['peso'], color='cyan')
plt.title('Histograma de peso', color='green')
plt.xlabel('Peso', color='pink')
plt.ylabel('Frecuencia', color='y')
plt.show()
```

```python
plt.hist(df['peso'], color='cyan')
plt.hist(df['edad'],color='r')
plt.hist(df['cadera'],color='k')

plt.title('Histograma de peso, edad, cadsera', color='gray')
plt.show()
```

```python
sns.displot(df['peso'],color='cyan')
plt.title('Histograma de peso', c='b', size=30, family='serif', style='italic', weight='bold', variant='small-caps')
plt.xlabel('Peso', color='black')
plt.ylabel('Count', color='black')
plt.show()
```

```python
covid=pd.read_csv('COVID_RESUMIDO.CSV')
covid
```

```python
plt.figure(figsize=(20,5))
total_confirmados=covid.loc[:,]
colors=np.random.rand(32)
plt.scatter(total_confirmados['Entidad'], total_confirmados['Población'], 
            s=total_confirmados['TotalConfirmadoEdo']/700, c=colors, cmap='rainbow', edgecolors='red', alpha=0.9)

plt.title('Gráficas de burbujas de mexicanos contagiados por COVID a finales de Ago 2021')
plt.xlabel('Estados')
plt.ylabel('Población')
plt.xticks(rotation=45)

plt.show()
```
